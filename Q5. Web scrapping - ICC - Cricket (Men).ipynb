```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.icc-cricket.com/rankings/mens/team-rankings/odi'
```

```python
page = requests.get(url)
page
```

```python
page.content
```

```python
soup =BeautifulSoup(page.content,'html.parser')
print(soup.prettify())
```

```python
team = soup.find_all('span',class_='u-hide-phablet',limit = 10)
team
```

```python
Men_team = []
for m in team:
    m= m.get_text().replace('\n','')
    Men_team.append(m)
Men_team
```

```python
len(Men_team)
```

```python
match = soup.find_all('td',class_='rankings-block__banner--matches')
match
```

```python
Match_ = []
for i in match:
    i= i.get_text().replace('\n','')
    Match_.append(i)
Match_
```

```python
match_1 = soup.find_all('td',class_='table-body__cell u-center-text',limit=20)
match_1
```

```python
Full_info =[]
for i in match_1:
    Full_info.append(i.text)
Full_info

for i in range(0,len(Full_info),2):
    Match_.append(Full_info[i])
Match_
```

```python
Match_10 = pd.Series(Match_[:10])
Match_10
```

```python
len(Match_10)
```

```python
scrap_point=soup.find_all('td',class_='rankings-block__banner--points')
scrap_point
```

```python
point_pa = []

for i in scrap_point:
    i = i.get_text().replace('\n','')
    point_pa.append(i)
point_pa
```

```python
point_info = soup.find_all('td',class_='table-body__cell u-center-text')
point_info
```

```python
P_info =[]
for j in point_info:
    P_info.append(j.text)
P_info

for j in range(1,len(P_info),2):
    point_pa.append(P_info[j])
point_pa
```

```python
new_point = point_pa[:10]
new_point
```

```python
len(new_point)
```

```python
scrap_rating = soup.find_all('td',class_='rankings-block__banner--rating u-text-right')
scrap_rating
```

```python
rating_single = []
for rs in scrap_rating:
    rating_single.append(rs.text.strip())
rating_single
```

```python
scrap_rating_full = soup.find_all('td',class_='table-body__cell u-text-right rating')
scrap_rating_full
```

```python
complete_rating = []

for cr in scrap_rating_full:
    complete_rating.append(cr.text)
complete_rating

```

```python
final_rating = rating_single[:1]+complete_rating[:9]
pd.Series(final_rating)
```

```python
len(final_rating)
```

```python
ODI_Match_men = pd.DataFrame({'Team':Men_team,'Matches':Match_10,'Point':new_point,'Rating':final_rating})
ODI_Match_men
```

### 2. Top 10 ODI Batsmen in men along with the records of their team and rating.

```python
url = 'https://www.icc-cricket.com/rankings/mens/player-rankings/test/batting'
```

```python
page = requests.get(url)
page
```

```python
page.content
```

```python
soup1 = BeautifulSoup(page.content,'html.parser')
print(soup1.prettify())
```

```python
scrap_batsmen = soup1.find_all('div',class_='rankings-block__banner--name-large')
scrap_batsmen
```

```python
batsmen = []
for i in scrap_batsmen:
    batsmen.append(i.text.strip())
batsmen
```

```python
scrap_batsmen_full = soup1.find_all('td',class_='table-body__cell rankings-table__name name')
scrap_batsmen_full
```

```python
Team = []

for t in scrap_batsmen_full:
    t=t.get_text().replace('\n','')
    Team.append(t)
Team
```

```python
Final_Batsmen = batsmen[:1]+Team[:9]
Final_Batsmen
```

```python
len(Final_Batsmen)
```

```python
scrap_team = soup1.find_all('div',class_='rankings-block__banner--nationality')
scrap_team
```

```python
team1 = []

for a in scrap_team:
    team1.append(a.text.strip())
team1
```

```python
scrap_rest_team = soup1.find_all('td',class_='table-body__cell nationality-logo rankings-table__team')
scrap_rest_team
```

```python
Team_men = []
for i in scrap_rest_team:
    Team_men.append(i.text.replace('\n',''))
Team_men
    
```

```python
final_team = team1[:1]+Team_men[:9]
pd.Series(final_team)
```

```python
len(final_team)
```

```python
scrap_rating = soup1.find_all('div',class_="rankings-block__banner--rating")
scrap_rating
```

```python
rating1 = []
for r in scrap_rating:
    rating1.append(r.text.strip())
rating1
```

```python
scrap_rating_rest = soup1.find_all('td',class_='table-body__cell rating',limit = 9)
scrap_rating_rest
```

```python
rating_rest = []
for rr in scrap_rating_rest:
    rating_rest.append(rr.text.strip())
pd.Series(rating_rest)
```

```python
final_rating = rating1[:1]+rating_rest[:]
final_rating
```

```python
len(final_rating)
```

```python
Top_10_batsmen = pd.DataFrame({'Batsmen in Men':Final_Batsmen,'Team':final_team,'Rating':final_rating})
Top_10_batsmen
```
