```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.nobroker.in/property/sale/hyderabad/Kundanbagh%20Colony?searchParam=W3sibGF0IjoxNy40Mzc1MDg3LCJsb24iOjc4LjQ1Njk0MDgsInBsYWNlSWQiOiJDaElKUDdvYlhMZVF5enNSTFBJR1BnSE1iQkUiLCJwbGFjZU5hbWUiOiJCZWd1bXBldCBNZXRybyBTdGF0aW9uIn1d&radius=2.0'
```

```python
page = requests.get(url)
page
```

```python

soup =BeautifulSoup(page.content,'html.parser')
print(soup.prettify())
```

```python
title = soup.find_all('h2',class_='heading-6 font-semi-bold nb__1AShY')
title
```

```python
Title = []

for t in title:
    t = t.get_text().replace('\n','')
    Title.append(t)
Title
```

```python
len(Title)
```

```python
location = soup.find_all('div',class_='nb__2CMjv')
location
```

```python
Loc = []
for i in location:
    i=i.get_text().replace('\n','')
    Loc.append(i)
Loc
```

```python
len(Loc)
```

```python
emi = soup.find_all('div',class_='font-semi-bold heading-6')
emi
```

```python
EMI_e = []
for e in emi:
    e = e.get_text().replace('\n','')
    EMI_e.append(e)
EMI_e
```

```python
#EMI
emi_info = []
for em in range(1,len(EMI_e),3):
    emi_info.append(EMI_e[em])
emi_info
```

```python
len(emi_info)
```

```python
#Area 
area_info = []
for ar in range(0,len(EMI_e),3):
    area_info.append(EMI_e[ar])
area_info
```

```python
len(area_info)
```

```python
#Price 
price_info = []
for pr in range(2,len(EMI_e),3):
    price_info.append(EMI_e[pr])
price_info
```

```python
len(price_info)
```

```python
Housing_info = pd.DataFrame({'Title':Title,'Location':Loc,'Area':area_info,'EMI':emi_info,'Price':price_info})
Housing_info
```
